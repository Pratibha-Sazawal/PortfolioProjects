```python
# Read in the data

# pd.read_csv("C:\Users\prati\Desktop\movies.csv") -- get an error
df = pd.read_csv(r"C:\Users\prati\Desktop\movies.csv")
```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
    
# Drop nulls

df = df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns  -- Don't want decimal place in budget and gross columns

df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
```

```python
# year and released should match

pd.read_csv(r"C:\Users\prati\Desktop\movies.csv", usecols = ["year", "released"])
```

```python
# Create correct year column

# We will extract the year from the released column and create a new column
# Need to convert released column type to string first

df["yearcorrect"] = df["released"].astype(str).str.split(",").str[-1].astype(str).str[1:5]
df
```

```python
# Sorting movies by gross revenue

df = df.sort_values(by = ["gross"], inplace = False, ascending= False)
```

```python
# Let's start looking at correlation

df.corr()   # This correlation works only on numerical fields - Pearson correlation

# Different types of correlation - pearson, kendall, spearman - pearson is default - all give slightly different results

# Budget and gross have a high correlation
```

```python
correlation_matrix = df.corr(method = "pearson")

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# Looking at company

df.head()
# Company is not numeric, but we can create a numeric representation of it
```

```python
df_numerised = df

for col_name in df_numerised.columns:
    if df_numerised[col_name].dtype == "object":   
# If column is name, genre, company, country etc. -- we want to change it to a category type
        df_numerised[col_name] = df_numerised[col_name].astype("category")
        df_numerised[col_name] = df_numerised[col_name].cat.codes   # gives the categories the random numerisation
        
df_numerised
```

```python
correlation_matrix = df_numerised.corr(method = "pearson")

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()

# Now we have every single field 
```

```python
df_numerised.corr()

# We want to filter it and see the ones with the highest correlation
```

```python
# Unstacking

correlation_mat = df_numerised.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs

# Unstacks everything and gives all the correlations for each column
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs

# we get sorted pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr

# Company doesn't have a high correlation with gross
# But votes and budget have the highest correlation to gross
```
